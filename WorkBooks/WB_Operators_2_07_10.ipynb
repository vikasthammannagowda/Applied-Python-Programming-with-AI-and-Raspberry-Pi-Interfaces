# Assignment Operators

<font size="4">Assignment Operators are used to assign values to variables. This operator is used to assign the value of the right side of the expression to the left side operand.</font>.

Example:

num_1 = 10

num_2 = 20

num_result = num_1 + num_2

Here, python first computes the operation on the Right Hand Side (RHS) i.e., num_1 + num_2 and then the values is assigned to the variable on Left Hand Side (LHS).


# Arithmetic Assignment Operator

The Arithmetic Assignment Operator is used to perform an arithmetic aperation of the right-hand side operand with the left-hand side operand and then assigning the result to the left operand.

Example: You are playing a game and have just found a chest containing 1000 gold coins. You currently possess 350 coins. How will you update your total?

Solution: You add the current coins with the coins you discovered (this is the RHS part) and assign this value back to the current coins.



```python
current_coins = 350 
discovers_coins = 1000

current_coins = current_coins + discovers_coins
# RHS: 
# LHS:
print("Current coins:", current_coins)
```

```python
current_coins = 350 
discovers_coins = 1000

current_coins +=  discovers_coins # current_coins = current_coins + discovers_coins
print("Current coins:", current_coins)
```

```python
num = 10
num *= 2 # num = num * 2
print(num)
```

# Special Operators

**in** and **not in** are used to test whether a value or variable is in a sequence.

**in**: True if value is found in the sequence

**not in**:  True if value is not found in the sequence
