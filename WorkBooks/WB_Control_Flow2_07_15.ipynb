# Problem: Check if an input year is a leap year or not

Every year that is exactly divisible by four is a leap year, 

except for years that are exactly divisible by 100, 

but these centurial years are leap years if they are exactly divisible by 400. 

```python
# 1700 not a leap year
# 1800, 1900, not a leap
#1600, 2000 : leap year
```

Divisibility by 4: The year must be divisible by 4. If a year is not divisible by 4, it is not a leap year.


Not Divisible by 100: If the year is divisible by 4, then check if it is also divisible by 100. If it is divisible by 100, it is not a leap year.


Divisibility by 400: If the year is divisible by 100, then check if it is also divisible by 400. If it is divisible by 400, it is a leap year. If it is not divisible by 400, it is not a leap year.

```python
year = int(input("Please enter a year:"))

if ((year % 4) == 0): # 2024 % 4 = 0, 1700 % 4 = 0, 1600 % 4 = 0
    # proceed to next step
    if ((year % 100) == 0): # 2024 % 100 = 24, 1700 % 100 = 0, 1600 % 100 = 0
        # proceed to next step
        if ((year % 400) == 0): # 1700 % 400 = 100, 1600 % 400 = 0
            print("Leap year")
        else:
            print("Not a leap year!")
    else:
        print("Leap year")
else:
    print("Not a leap year!")
    
```

# Pay calculator

<font size="4">Write a Python program that calculates the total weekly pay for an employee based on their hourly wage and total hours worked, where any hours over 40 are paid at 1.5 times the regular hourly rate. </font>

```python
# Prompt the user to input the hourly wage
pay_per_hour = float(input("Please enter the pay per hour:"))

# Prompt the user to input the total number of hours worked in a week
hours_worked = float(input("Please enter the total hours worked in a week"))

# Check for the total hours worked
if (hours_worked <= 40): # if it is less than or equal to 40 hours then normal pay
    # Calculate the total pay
    total_pay = hours_worked * pay_per_hour
else: # means, the employee has worked more than 40 hours
    # Calculate regular pay for the first 40 hours
    regular_pay = 40 * pay_per_hour
    # Calculate overtime pay
    over_time = hours_worked - 40
    over_time_pay = over_time * pay_per_hour * 1.5
    # Total pay is the sum of regular pay and overtime pay
    total_pay = regular_pay + over_time_pay
    

# Print the total pay
print("The total pay for the week is:", total_pay)
```
