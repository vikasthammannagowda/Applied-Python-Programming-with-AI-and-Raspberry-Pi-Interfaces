# For loops

A **for loop** is basically used to iterate over a set of elements.

These elements are generally a range of values or a list of values.

Example: Say you want to display all even numbers between 1 and 11. 

If we use a while loop, we first initialize a variable to 1, check and display if the number is even until the number reaches 11, i.e.,

```python
num_1 = 1
while (num_1 <= 11):
    if((num_1 % 2) == 0):
        print(num_1)
    num_1 += 1
```

```python
round(200/7.2, 3)
```

Instead of this, we can generate a range of values from 1 to 11 in increments of 1, then check each of these numbers and display the even numbers.

We use a function to generate those values. A **range()** function generates a sequence of integers between the two
numbers given a step size. This integer sequence is inclusive of the start and exclusive of the end of the
sequence.

Note: A function is like a small piece of code that is inbuilt into the interpreter. It takes some values as input/s and generates outputs.



```python
min(10,20)
```

```python
max(10, 20)
```

# Problem 1 

<font size="4">Find the sum to first "n" natural numbers using a for loop</font>

```python
sum = 0
n_th_number = int(input("Enter the largers natural number:"))
for num in range(1, (n_th_number+1)):
    #print(num)
    sum = sum + num # itr 1: sum = 0 + 1, itr2: sum = 1 + 2

print("The sum is:", sum)
```

Check if an input integer is a prime number.

Logic: A prime number is a positive integer that is greater than 1 and is divisible by 1 and itself

Python Logic: I take a number and check for its divisiblity with each value 2 to num-1 (*), even if the number is divisible by one of these values, it is not a prime number else, it is a prime number

num = 35

itr4: 35 % 5 == 0 -> True

```python
agent = True
num = int(input("Please enter a number to check if its a prime or composite")) # 99
for i in range (2, num): # (2, 3, 4, 5, ......, num-1)
    if(num % i == 0): # 99 % 2 == 0 -? 99 % 3 == 0
        agent = False
        break

if(agent == True):
    print(num, "is a prime number")
else:
    print(num, "is a composite number")
    
```
